```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sps
import seaborn as sns

from lib.ml import dist
```

```python
estimator = dist.ZeroInflattedPoisson()

N = 10000
data = sps.poisson.rvs(10, size=N)
data[sps.bernoulli.rvs(0.5, size=N).astype(bool)] = 0
estimator.fit(data)

plt.figure(figsize=(12, 4))

plt.hist(data, bins=21, range=(-0.5, 20.5), density=True, label='Data')

grid = np.arange(0, 20)
plt.scatter(grid , estimator.proba(grid), color='red', label='Distribution estimation', zorder=10)

plt.legend(fontsize=14)

plt.show()
```
